Display the row that contains the ‘Model’ of ‘Mazda RX4’

```python
Montales_Pandas.loc[Montales_Pandas['Model']=='Mazda RX4', ['Model']]
```

How many cylinders (‘cyl’) does the car model ‘Camaro Z28’ have?

```python
Montales_Pandas.loc[Montales_Pandas['Model']=='Camaro Z28', ['Model','cyl']]
```

Determine how many cylinders (‘cyl’) and what gear type (‘gear’) do the car models ‘Mazda RX4
Wag’, ‘Ford Pantera L’ and ‘Honda Civic’ have.


```python
Montales_Pandas.loc[Montales_Pandas['Model'].isin(['Mazda RX4 Wag','Ford Pantera L','Honda Civic']), ['Model','cyl', 'gear']]
```
